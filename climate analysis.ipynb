```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(daily_data['DailyAverageDryBulbTemperature'], bins=20, kde=True, color='blue')
plt.title('Distribution of Daily Average Temperature')
plt.xlabel('Temperature (°F)')
plt.ylabel('Frequency')
plt.show()
```

```python
numeric_columns = hourly_data.select_dtypes(include=['float64', 'int64']).columns
correlation_matrix = hourly_data[numeric_columns].corr()
plt.figure(figsize=(20, 15))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix for Hourly Data')
plt.show()
```
